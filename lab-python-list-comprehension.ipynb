## Exercise: Managing Customer Orders Optimized with Comprehension

````python
def manage_orders():
    def initialize_inventory(products):
        inventory = {}

        for product in products:
            quantity = input(f"Enter quantity of {product}: ").strip()
            
            
            while not quantity.isdigit():
                print("Please enter a number", "\n")
                quantity = input(f"Enter quantity of {product}: ").strip()

            quantity = int(quantity)
            inventory[product] = quantity
                
        return inventory

    def get_customer_orders(inventory):
        
        customer_orders = {product:0 for product in inventory.keys()}
        continue_buying = 'yes'

        while continue_buying == 'yes':
            print(f'Available products: {','.join(inventory.keys())}',"\n")
            product = input('Which product do you want to buy?:')
            if product in inventory.keys():
                while True:
                    try:
                        order = int(input(f'Enter the quantity of the product {product} to purchase: ').strip().lower())
                    
                        if order < 0:
                            print(f"Invalid input. Quantity has to be grater than 0.")
                        elif order > inventory[product]:
                            print((f'Not enought stock! Only {inventory[product]} available'))
                        else:
                            customer_orders[product] += order
                            break

                    except ValueError:
                        print('Invalid input. Quantity has to be a number')
            else:
                print(f"Sorry we don't have {product} in our inventory.")
                continue
            
            continue_buying = input("Do you want to buy more items? (yes/no): ").strip().lower()
            while continue_buying not in ("no", "yes"):
                print(f"'{continue_buying}' is not a correct option.")
                continue_buying = input("Type Yes or No: ").strip().lower()
            
        # print("------Customer´s order------")
        # print (f"The customer has ordered {customer_orders}", "\n")

        return customer_orders

    def update_inventory(customer_orders,inventory):

        updated_inv = {product:value - customer_orders.get(product,0) for product,value in inventory.items()}
        
        # for item in customer_orders:
        #     inventory[item] -= 1
        #     if inventory[item] == 0:
        #         del inventory[item]

        out_stock= [product for product, value in updated_inv.items() if value == 0]
        
        return updated_inv,out_stock
    
    def calculate_order_statistics(customer_orders,products):

        total_products_ordered = len ([product for product,quantity in customer_orders.items() if quantity > 0])
        percentage_ordered = round((total_products_ordered/len(products))*100,1)

        return total_products_ordered, percentage_ordered

    def print_order_statistics(order_statistics):

        total_products_ordered, percentage_of_products_ordered = order_statistics
        print("------Order Statistics------")
        print(f"Total products ordered: {total_products_ordered}")
        print(f"Percentage of products ordered: {percentage_of_products_ordered}%","\n")

    def print_updated_inventory(updated_inv,out_stock):
        print("------Updated inventory------")
        for key,value in updated_inv.items():
            print(f"{key}: {value}")
        print()
        print ("------Products out of stock------")
        print(out_stock)

    def total_price(customer_orders):

        prices_dict= {item: float(input(f"Enter the price of one {item}: ").replace(",",".")) for item, value in customer_orders.items() if value > 0}

        for key,value in prices_dict.items():
            print(f"The price of one {key} is: {value}$")
        
        sum_prices = sum(prices_dict.values())
        print('Total price of the customer order:',sum_prices,"$")
        print()

    products = ["t-shirt", "mug", "hat", "book", "keychain"]

    inventory = initialize_inventory(products) 
    customer_orders = get_customer_orders(inventory)
    updated_inv, out_stock = update_inventory(customer_orders,inventory)
    order_statistics = calculate_order_statistics(customer_orders,products)
    print_order_statistics(order_statistics)
    print_updated_inventory(updated_inv,out_stock)
    print()
    total_price(customer_orders)
    
    
````
